```python
# -*- coding: utf-8 -*-
import cv2
from ultralytics import YOLO
import numpy as np

# Load a pre-trained YOLOv8n pose detection model
model = YOLO('yolov8n-pose.pt')

# Specify the video source (ensure this path is correctly set for your environment)
source = r"C:\\Users\\Asus Np\\Videos\\VideoProc Converter AI\\GOPR4094_2.MP4"

# Initialize video capture from the video file
cap = cv2.VideoCapture(source)

# Check if the video opened successfully
if not cap.isOpened():
    print("Error: Could not open the video.")
    exit()

all_keypoints = []  # List to store keypoints for all frames
frame_count = 0  # Frame counter

# Process each frame of the video
while cap.isOpened():
    # Read a frame from the video
    ret, frame = cap.read()

    # Break the loop if no more frames are available
    if not ret:
        break

    # Run YOLO pose detection on the current frame
    results = model.predict(frame, show=False, conf=0.3, save=False)

    # Process the detected keypoints for each result in the frame
    for result in results:
        for keypoint_indx, keypoint in enumerate(result.keypoints.xy[0]):
            keys = (int(keypoint_indx), int(keypoint[0]), int(keypoint[1]))
            all_keypoints.append(keys)
    
    frame_count += 1
    print(f"Processed frame: {frame_count}")

# Release the video capture object and close any OpenCV windows
cap.release()
cv2.destroyAllWindows()

# Now, let's extract specific keypoints and calculate the angle between them
if len(all_keypoints) > 12:
    x1, y1 = all_keypoints[5][1], all_keypoints[5][2]
    x2, y2 = all_keypoints[7][1], all_keypoints[7][2]
    x3, y3 = all_keypoints[11][1], all_keypoints[11][2]

    dot1 = np.array([x1, y1])
    dot2 = np.array([x2, y2])
    dot3 = np.array([x3, y3])

    # Calculate vectors representing two lines formed by keypoints
    vector1 = dot2 - dot1
    vector2 = dot3 - dot1

    # Compute the dot product and magnitudes of vectors
    dot_product = np.dot(vector1, vector2)
    magnitude1 = np.linalg.norm(vector1)
    magnitude2 = np.linalg.norm(vector2)

    # Calculate the angle in radians, then convert to degrees
    angle_radians = np.arccos(dot_product / (magnitude1 * magnitude2))
    angle_degrees = np.degrees(angle_radians)

    print(f"Calculated Angle (degrees): {angle_degrees:.2f}")
else:
    print("Insufficient keypoints detected to calculate angles.")
```
